Let's create the bronze layer to land data

```python
%%sql
CREATE TABLE IF NOT EXISTS wwi_bronze.sales_orders (
  OrderID BIGINT,
  CustomerID BIGINT,
  SalespersonPersonID BIGINT,
  OrderDate STRING,
  ExpectedDeliveryDate STRING,
  CustomerPurchaseOrderNumber STRING,
  IsUndersupplyBackordered STRING,
  Comments STRING,
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```

```python
%%sql

CREATE TABLE IF NOT EXISTS wwi_bronze.sales_order_lines (
  OrderLineID BIGINT,
  OrderID BIGINT,
  StockItemID BIGINT,
  Quantity BIGINT,
  UnitPrice DECIMAL(18,2),
  TaxRate DECIMAL(9,4),
  PickedQuantity BIGINT,
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```

```python
%%sql

CREATE TABLE IF NOT EXISTS wwi_bronze.sales_customers (
  CustomerID BIGINT,
  CustomerName STRING,
  BillToCustomerID BIGINT,
  CustomerCategoryID BIGINT,
  PrimaryContactPersonID BIGINT,
  DeliveryCityID BIGINT,
  PostalCityID BIGINT,
  CreditLimit DECIMAL(18,2),
  AccountOpenedDate STRING,
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```

```python
%%sql
CREATE TABLE IF NOT EXISTS wwi_bronze.warehouse_stock_items (
  StockItemID BIGINT,
  StockItemName STRING,
  SupplierID BIGINT,
  Brand STRING,
  Size STRING,
  UnitPrice DECIMAL(18,2),
  TaxRate DECIMAL(9,4),
  TypicalWeightPerUnit DECIMAL(18,4),
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```

```python
%%sql
CREATE TABLE IF NOT EXISTS wwi_bronze.purchasing_suppliers (
  SupplierID BIGINT,
  SupplierName STRING,
  SupplierCategoryID BIGINT,
  PrimaryContactPersonID BIGINT,
  DeliveryCityID BIGINT,
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```

```python
%%sql
CREATE TABLE IF NOT EXISTS wwi_bronze.application_cities (
  CityID BIGINT,
  CityName STRING,
  StateProvinceID BIGINT,
  CountryID BIGINT,
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```

```python
%%sql
CREATE TABLE IF NOT EXISTS wwi_bronze.application_countries (
  CountryID BIGINT,
  CountryName STRING,
  FormalName STRING,
  IsoAlpha3Code STRING,
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```

```python
%%sql
CREATE TABLE IF NOT EXISTS wwi_bronze.sales_invoices (
  InvoiceID BIGINT,
  CustomerID BIGINT,
  InvoiceDate STRING,
  OrderID BIGINT,
  BillToCustomerID BIGINT,
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```

```python

%%sql
CREATE TABLE IF NOT EXISTS wwi_bronze.sales_invoice_lines (
  InvoiceLineID BIGINT,
  InvoiceID BIGINT,
  StockItemID BIGINT,
  Quantity BIGINT,
  UnitPrice DECIMAL(18,2),
  TaxRate DECIMAL(9,4),
  LineProfit DECIMAL(18,2),
  _ingested_at TIMESTAMP,
  _source_file STRING
) USING DELTA;
```
